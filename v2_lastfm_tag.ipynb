**TAGS LASTFM: DADOS DE 2018**

```python
import pandas as pd
import requests
import time

API_KEY = '0ca4b754cfeaf45d3bb6d4e8b9351e7e'  # Replace with your Last.fm API key
BASE_URL = 'http://ws.audioscrobbler.com/2.0/'

# Load your CSV
df = pd.read_excel('/content/base_hot100_2018.xlsx')  # Change to your actual CSV filename

# Prepare a list to hold genres
genres = []

# Loop over each artist and fetch the top tag
for artist in df['main_artist']:
    try:
        params = {
            'method': 'artist.getTopTags',
            'artist': artist,
            'api_key': API_KEY,
            'format': 'json'
        }
        response = requests.get(BASE_URL, params=params)
        data = response.json()

        # Get the first tag if available
        tags = data.get('toptags', {}).get('tag', [])
        if isinstance(tags, list) and tags:
            top_tag = tags[0].get('name', '')
        elif isinstance(tags, dict):
            # Sometimes the API returns a dict instead of list if only one tag exists
            top_tag = tags.get('name', '')
        else:
            top_tag = ''

        genres.append(top_tag)

    except Exception as e:
        print(f"Error processing artist '{artist}': {e}")
        genres.append('')  # Append empty if there's an error

    time.sleep(0.25)  # Be nice to the API, avoid getting rate-limited

# Add the genre column
df['genre'] = genres

# Save to a new CSV
df.to_excel('tag_lastfm_2018.xlsx', index=False)
print("Finished! Saved as 'tag_lastfm_2018.xlsx'")
```

**MERGE DAS PLANILHAS**

```python
import pandas as pd

# Lista com os anos que vocÃª quer juntar
anos = range(2018, 2024)  # 2018 a 2023 inclusive

# Lista para armazenar os DataFrames
dfs = []

for ano in anos:
    arquivo = f"/content/tag_lastfm_{ano}.xlsx"
    df = pd.read_excel(arquivo)
    dfs.append(df)

# Concatenar todos os DataFrames
df_merged = pd.concat(dfs, ignore_index=True)

# Salvar em um novo arquivo Excel
df_merged.to_excel("tag_lastfm_full_merged.xlsx", index=False)
```
