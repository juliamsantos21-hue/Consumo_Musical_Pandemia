**HOT 100, ANO: 2018**

```python
import billboard
import csv
from datetime import date, timedelta
```

```python
# Função para gerar todas as datas de sábado em 2018 (datas dos charts)
def get_chart_dates(year):
    d = date(year, 1, 6)  # primeiro sábado de 2018
    dates = []
    while d.year == year:
        dates.append(d.isoformat())
        d += timedelta(days=7)
    return dates

dates_2018 = get_chart_dates(2018)

# Criar e abrir o CSV
with open("billboard_hot100_2018.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["rank", "title", "artist", "date"])  # cabeçalho

    # Percorrer todas as semanas de 2018
    for week in dates_2018:
        print(f"Baixando dados da semana {week}...")
        chart = billboard.ChartData("hot-100", date=week)

        # Gravar cada faixa (1–100)
        for entry in chart:
            writer.writerow([entry.rank, entry.title, entry.artist, week])

print("✅ CSV gerado: billboard_hot100_2018.csv")
```
