```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import time

# --- escolha o ano aqui ---
ano = 2018

df = pd.read_excel("/content/hot100_base_artista_limpo2018.xlsx")

# --- autentica√ß√£o Spotify ---
client_id = "654ce0a7c6664a84aca1e4d99406e15b"
client_secret = "c38902f4e99746ffb02756abda23c4bc"

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(
        client_id=client_id,
        client_secret=client_secret
    )
)

# --- garantir datetime ---
df['date'] = pd.to_datetime(df['date'])

# --- filtra ano escolhido ---
df_year = df[df['date'].dt.year == ano].copy()

# --- inicializa colunas novas ---
df_year['track_id'] = None
df_year['duration_mmss'] = None
df_year['album_name'] = None
df_year['album_release_type'] = None
df_year['release_date_prioridade'] = None
df_year['release_date_minima'] = None
df_year['valida_para_billboard'] = None
df_year['query_usada'] = None
df_year['fallback_usado'] = "N√£o"

# --- fun√ß√£o para converter minutos decimais para mm:ss ---
def min_to_mmss(duration):
    if pd.isna(duration):
        return None
    minutes = int(duration)
    seconds = int((duration - minutes) * 60)  # truncar em vez de arredondar
    return f"{minutes}:{seconds:02d}"

# --- fun√ß√£o para normalizar release_date ---
def normaliza_release_date(release_date):
    if release_date is None or pd.isna(release_date):
        return None
    if len(release_date) == 4:  # s√≥ ano
        return f"{release_date}-01-01"
    elif len(release_date) == 7:  # ano-m√™s
        return f"{release_date}-01"
    else:  # j√° vem completa
        return release_date

# --- loop para preencher as colunas ---
for idx, row in df_year.iterrows():
    track_name = str(row['title']).strip()
    artist_name = str(row['artista_principal']).strip()
    week_date = row['date']

    # primeira query (mais precisa)
    query_precisa = f'track:"{track_name}" artist:"{artist_name}"'
    # fallback (s√≥ pelo t√≠tulo)
    query_fallback = f'track:"{track_name}"'

    chosen_track = None
    candidatos = []

    try:
        # --- primeira tentativa ---
        search_result = sp.search(q=query_precisa, type='track', limit=10)
        candidatos = search_result['tracks']['items']

        # --- fallback se nada encontrado ---
        if not candidatos:
            search_result = sp.search(q=query_fallback, type='track', limit=10)
            candidatos = search_result['tracks']['items']
            df_year.at[idx, 'fallback_usado'] = "Sim"
            query_usada = query_fallback
        else:
            query_usada = query_precisa

        validos = []
        for track in candidatos:
            release_date = normaliza_release_date(track['album']['release_date'])
            album_type = track['album'].get('album_type', 'unknown')

            release_date_dt = pd.to_datetime(release_date, errors='coerce')
            validos.append({
                "track": track,
                "release_date": release_date,
                "release_date_dt": release_date_dt,
                "album_type": album_type
            })

        chosen_release_dt_prioridade = None
        chosen_release_dt_minima = None

        if validos:
            # --- prioridade: single > album > compilation ---
            for tipo in ["single", "album", "compilation"]:
                candidatos_tipo = [
                    c for c in validos
                    if c["album_type"] == tipo and pd.notna(c["release_date_dt"]) and c["release_date_dt"] <= week_date
                ]
                if candidatos_tipo:
                    chosen = max(candidatos_tipo, key=lambda x: x["release_date_dt"])
                    chosen_track = chosen["track"]
                    chosen_release_dt_prioridade = chosen["release_date"]
                    break

            # se nenhum v√°lido, usa o primeiro como fallback
            if not chosen_track:
                chosen = validos[0]
                chosen_track = chosen["track"]
                chosen_release_dt_prioridade = chosen["release_date"]

            # --- menor data poss√≠vel (apenas <= week_date) ---
            valid_dates = [
                c["release_date_dt"]
                for c in validos
                if pd.notna(c["release_date_dt"]) and c["release_date_dt"] <= week_date
            ]
            if valid_dates:
                chosen_release_dt_minima = min(valid_dates)

        # --- preencher colunas ---
        if chosen_track:
            duration_min = round(chosen_track['duration_ms'] / 60000, 2)
            df_year.at[idx, 'track_id'] = chosen_track['id']
            df_year.at[idx, 'duration_mmss'] = min_to_mmss(duration_min)
            df_year.at[idx, 'album_name'] = chosen_track['album']['name']
            df_year.at[idx, 'album_release_type'] = chosen_track['album'].get('album_type', 'unknown')
            df_year.at[idx, 'release_date_prioridade'] = chosen_release_dt_prioridade
            df_year.at[idx, 'release_date_minima'] = chosen_release_dt_minima.strftime("%Y-%m-%d") if chosen_release_dt_minima else None
            df_year.at[idx, 'query_usada'] = query_usada

            if pd.notna(chosen_release_dt_prioridade) and pd.to_datetime(chosen_release_dt_prioridade) <= week_date:
                df_year.at[idx, 'valida_para_billboard'] = "Sim"
            else:
                df_year.at[idx, 'valida_para_billboard'] = "N√£o"

            print(f"‚úÖ Faixa encontrada: {track_name} ‚Äî {artist_name}")
        else:
            print(f"‚ùå Nenhum resultado para: {track_name} ‚Äî {artist_name}")

    except Exception as e:
        print(f"Erro na faixa {track_name} ‚Äî {artist_name}: {e}")

    time.sleep(0.2)  # pausa para rate limit

# --- salvar CSV final ---
output_name = f"hot100_{ano}_spotify_full.csv"
df_year.to_csv(output_name, index=False)
print(f"‚úÖ CSV criado: {output_name}")
```
